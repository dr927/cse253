```python
from os import listdir
from PIL import Image
import numpy as np


# The relative path to your CAFE-Gamma dataset
data_dir = "/Users/Derar/Downloads/CAFE/"

# Dictionary of semantic "label" to emotions
emotion_dict = {"h": "happy", "ht": "happy with teeth", "m": "maudlin",
    "s": "surprise", "f": "fear", "a": "anger", "d": "disgust", "n": "neutral"}


def load_data(data_dir):
        """ Load all PGM images stored in your data directory into a list of NumPy
        arrays with a list of corresponding labels.

        Args:
            data_dir: The relative filepath to the CAFE dataset.
        Returns:
            images: A list containing every image in CAFE as an array.
            labels: A list of the corresponding labels (filenames) for each image.
        """
        # Get the list of image file names
        all_files = listdir(data_dir)

        # Store the images as arrays and their labels in two lists
        images = []
        labels = []

        for file in all_files:
            # Load in the files as PIL images and convert to NumPy arrays
            img = Image.open(data_dir + file)
            images.append(np.array(img))
            labels.append(file)

        print("Total number of images:", len(images), "and labels:", len(labels))

        return images, labels



def display_face(img):
        """ Display the input image and optionally save as a PNG.

        Args:
            img: The NumPy array or image to display

        Returns: None
        """
        # Convert img to PIL Image object (if it's an ndarray)
        if type(img) == np.ndarray:
            print("Converting from array to PIL Image")
            img = Image.fromarray(img)

        # Display the image
        img.show()
```

```python
LA = np.linalg
def pca(images):
    mean_img = sum(images)/len(images)                         #mean of the emotion faces
    faces = [x-mean_img for x in images]                       #substract the mean from the faces
    face_arr = np.asarray(faces)
    face_arr = np.reshape(face_arr,(len(images),-1))
    face_arr = np.divide(face_arr,255.0)
    face_arr = np.divide(face_arr, np.std(face_arr,axis = 1)[:,None])
    face_arr = face_arr.T
    covariance = np.cov(face_arr.T)  #covariance matrix of the images
    w, v = LA.eigh(covariance)      #eigenvalues and eigenvectors of covariance matrix
    w, v = w[::-1], v[::-1]
    eig_vecs = np.matmul(face_arr, v)
    eig_vecs /= LA.norm(eig_vecs, axis=0)[None, :]
    w = np.reshape(w,(1,w.shape[0]))
    eig_vecs = np.divide(eig_vecs,np.sqrt(w))
    print eig_vecs[:,0]
    pc = np.reshape(eig_vecs,(380,240,len(images)))
    
    return pc 



pc = pca(images)
columns = 3                                        #plot the first 6 principal components of the faces
rows = 2
fig=plt.figure(figsize=(8, 8))    
for i in range(1, columns*rows +1 ):
    img = pc[:,:,i-1]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img, cmap='gray')
    
plt.show()
```

```python
def project(pc,images,k):
    [a,b,c] = pc.shape
    mean_img = sum(images)/len(images)
    images = [x-mean_img for x in images] 
    images = np.asarray(images)
    images = np.divide(images,255.0)
    images = np.reshape(images,(images.shape[0], a*b))
    pc = np.reshape(pc,(images.shape[0], a*b))
#     norm = LA.norm(pc,axis = 0)
#     pc = np.divide(pc,norm)
    principal = pc[0:k,:]
    reduced = np.matmul(principal,images.T)
    return reduced



def lr(set1,set2):
    
    row = set1.shape[0]
    column = set1.shape[1]
    learn_rate = 0.01
    s1 = []
    s2 = []
    holdout = []
    test = []
    e = []
    e_holdout = []
    
    for i in range(column):
        temp1 = np.append(1,(set1[:,i]))              #add a leading 1 to the data input to account for bias
        temp2 = np.append(1,(set2[:,i]))
        s1.append(temp1)
        s2.append(temp2)
        
    order = np.random.permutation(column)        #permute indices to choose random test and holdout subjects
    
    
    for i in range(len(order)):
        
        train1 = s1[:]
        train2 = s2[:]
        holdout.append(train1[order[i]])            #select holdout subject randomly
        holdout.append(train2[order[i]])
        del train1[order[i]]                        #delete the selected subjects for holdout and test from training
        del train2[order[i]]
        horder = np.random.permutation(len(train1))
        test.append(train1[horder[0]])
        test.append(train2[horder[0]])
        del train1[horder[0]]
        del train2[horder[0]]
        w, error,error_holdout = gd(train1,train2,learn_rate,holdout)
        e.append(error)
        e_holdout.append(error_holdout)
        #ploti(e, e_holdout)
        
        
def ploti(er,erh):
  
    er = np.vstack(er)
    erh = np.vstack(erh)
    epochs = range(1,11)
    train_loss = np.mean(er,axis = 0)
    hold_loss = np.mean(erh,axis =0)
    plt.plot(epochs,train_loss,'r--')
    plt.plot(epochs,hold_loss,'b-')
    plt.legend(['Training loss','Holdout loss'])
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Cross entropy loss')
    plt.show()
    
    
        
def cross_entropy(set1,set2,w):
    N = len(set1)
    error = 0
    for i in range(N):
        error  -= np.log(logistic(w,set1[i])) + np.log(1-logistic(w,set2[i]))
    error = error*(1/N)
    return error
    
    
    
    
def gd(set1,set2,alpha,holdout):                             #gradient descent update for a number of epochs

    bound = np.inf
    err = []                                                 #list to store the training error after each epoch
    err_h = []                                               #list to store the holdout error after each epoch
    epochs = 10
    w = np.zeros(set1[0].shape)                              #initialize the weights to zero
    w2 = np.zeros(set1[0].shape) 
    for s in range(epochs):
        sum1 = 0
        sum2 = 0
        for i in range(len(set1)):
            sum1+= alpha*(1-logistic(w,set1[i]))*set1[i]     #summation of the gradient term for n = 1 label
            sum2+= alpha*(0-logistic(w,set2[i]))*set2[i]     #summation of the gradient term for n = 0 label
        w2 = w2 + sum1 + sum2
         
        err.append(cross_entropy(set1,set2,w2))
        holdout_errorterm = cross_entropy(holdout[:1],holdout[1:],w2)
        err_h.append(holdout_errorterm)
        if holdout_errorterm < bound :
            bound = holdout_errorterm
            w = w2
    return w,err,err_h
    
    
    
def logistic(w,x):                                           # logistic discriminant product(prediction y)
    argument = -1*(np.matmul(w.T,x))
    y = 1 + np.exp(argument)
    y = 1/y
    return y
```
